# Functions Practical Question

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_evens(nums):
    return sum(x for x in nums if x % 2 == 0)

# Example
print(sum_of_evens([1, 2, 3, 4, 10]))  # 16
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(s):
    return s[::-1]

# Example
print(reverse_string("hello"))  # "olleh"
```

```python
3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.
def squares(nums):
    return [x * x for x in nums]

# Example
print(squares([1, 2, 3, 4]))  # [1, 4, 9, 16]
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2
    return True

# Example: primes from 1 to 200
primes_1_to_200 = [n for n in range(1, 201) if is_prime(n)]
print(primes_1_to_200)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1  # starts: 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        if self.count == 1:
            self.count += 1
            return 1
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return self.b

# Example
print(list(Fibonacci(7)))  # [0, 1, 1, 2, 3, 5, 8]
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for e in range(n + 1):
        yield 2 ** e

# Example
print(list(powers_of_two(5)))  # [1, 2, 4, 8, 16, 32]
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(path):
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            yield line.rstrip('\n')  # or just: yield line

# Example
# for line in read_lines("myfile.txt"):
#     print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [("a", 3), ("b", 1), ("c", 2)]
sorted_data = sorted(data, key=lambda t: t[1])
print(sorted_data)  # [('b', 1), ('c', 2), ('a', 3)]
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
temps_c = [0, 20, 37, 100]
temps_f = list(map(lambda c: c * 9 / 5 + 32, temps_c))
print(temps_f)  # [32.0, 68.0, 98.6, 212.0]
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = set("aeiouAEIOU")
    return ''.join(filter(lambda ch: ch not in vowels, s))

# Example
print(remove_vowels("Hello World"))  # "Hll Wrld"
```
