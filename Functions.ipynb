# Functions

1. What is the difference between a function and a method in Python?
-  Function: A block of code defined using def that performs a specific task/ computation.
-  Method: A function that is associated with an object/class. It is called on an instance (object) and often uses self as its first parameter.
  
2. Explain the concept of function arguments and parameters in Python.
-  Parameter: Variable listed inside the parentheses when defining a function.
-  Argument: Value passed to the function when calling it.

```python
# 3. What are the different ways to define and call a function in Python?
# Regular function:
def square(x):
    return x * x
print(square(5))
# Function with default arguments:
def greet(name="Guest"):
    print("Hello", name)
greet()  # uses default

# Function with keyword arguments:
def info(name, age):
    print(name, age)
info(age=20, name="Alice")

# Lambda function (anonymous):
double = lambda x: x * 2
print(double(4))

```

6. Explain the concept of generators in Python and how they are defined.
-  A generator is a special function that yields values one at a time using yield instead of return.
-  They produce values lazily (on demand).
7. What are the advantages of using generators over regular functions?
-  Memory efficient (don’t store all values at once).
-  Lazy evaluation (compute values only when needed).
-  Convenient for large data streams, files, or infinite sequences.
8. What is a lambda function in Python and when is it typically used?
-  An anonymous, single-expression function defined with lambda.
-  Used for short, throwaway functions (often with map, filter, or sorting).
9. Explain the purpose and usage of the `map()` function in Python.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
-  map(function, iterable) → Applies function to each element.
-  filter(function, iterable) → Keeps only elements for which function returns True.
-  reduce(function, iterable) (from functools) → Reduces iterable to a single value by applying function cumulatively.
